```python
# import the required libraries
from webbrowser import Chrome
from bs4 import BeautifulSoup as soup
import urllib.request as ureq
from urllib.request import urlopen
```

```python
# specify the url
url = 'https://clutch.co/web-developers'
```

```python
#Query the website and return the html to the variable 'page'
req = ureq.Request(url, headers={'User-Agent': 'Chrome/71.0'})
opener = ureq.build_opener()
response = opener.open(req)
page = response.read()
```

```python
# end the connection as we have html
response.close()
```

```python
# parse the html
page_soup = soup(page, "html.parser")
containers = page_soup.find_all("li", {"class:", "provider-row"})
print(len(containers))
```

```python
# collecting the info of the companies listed
container = containers[0]

company = container.findAll("h3", {"class:", "company-name"})
print(company[0].text)

ratings = container.findAll("div", {"class:", "provider-profile-rating-widget"})
print(ratings[0].text)
```

```python
# create a csv file to store the collected data
filename = "companies_data.csv"
f = open(filename,"w")
```

```python
column_names = "company,rating,reviews"
f.write(column_names)
```

```python
for container in containers:
    company = container.findAll("h3", {"class:", "company-name"})
    #print(company[0].text)

    ratings = container.findAll("div", {"class:", "provider-profile-rating-widget"})
    #print(ratings[0].text)
    
    rating = ratings[0].text.split(' ')
    rating = rating[0].split('\n\n\n')
        
    review = rating[1]
    
    print(company[0].text+","+rating[0]+","+review+"\n")
    f.write(company[0].text+","+rating[0]+","+review+"\n")
```
